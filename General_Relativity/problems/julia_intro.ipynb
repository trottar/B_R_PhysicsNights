```julia
import Pkg
# Testing basic Julia mathematical package
Pkg.add("Gadfly")
using Gadfly
# Type in greek letter and hit <tab> to use greek letter variables!
# eg.
# \delta <tab> will result in
δ = sqrt(complex(-1))
```

```julia
# You can call python packages right in Julia
Pkg.add("PyCall") # Note that capitalization matters
using PyCall
m = pyimport("math") # import python math package
print(m.exp(log(2)))
```
